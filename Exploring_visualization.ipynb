#**Task : Exploring and Visualizing a Simple Dataset**:

 Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.
Dataset:
 Iris Dataset (CSV format, can be loaded via seaborn or downloaded)

**Instructions:**

* Load the dataset using pandas.

* Print the shape, column names, and the first few rows using .head().

* Use .info() and .describe() for summary statistics.

**Visualize the dataset:**

* Create a scatter plot to show relationships between features.

* Use histograms to show value distributions.

* Use box plots to identify outliers.

* Use matplotlib and seaborn for plotting.


```python
# Import necessary libraries
import numpy as np                      # For numerical operations
import pandas as pd                     # For data handling and analysis
import seaborn as sns                   # For advanced data visualization
import matplotlib.pyplot as plt         # For plotting graphs

# Import Google Drive module and mount drive
from google.colab import drive
drive.mount('/content/drive')           # Connect Colab with Google Drive

# Load dataset from Google Drive
df = pd.read_csv('/content/drive/MyDrive/iris_dataset.csv',
                 names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'],  # Column names
                 header=None)           # No header in CSV file

# ---------------- Basic Inspection ----------------
print("Shape:", df.shape)              # Display number of rows and columns
print("\nColumns:", df.columns)        # Display column names
print("\nHead:\n", df.head())          # Show first 5 rows of dataset

print("\nInfo:")
print(df.info())                       # Show data types and non-null values

print("\nDescribe:\n", df.describe())  # Statistical summary of numerical columns

# ---------------- Scatter Plots ----------------

plt.figure(figsize=(8,6))              # Set figure size
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')  # Plot sepal length vs width
plt.title("Sepal Length vs Sepal Width")  # Title of graph
plt.grid(alpha=0.3)                   # Add grid with transparency
plt.tight_layout()                    # Adjust layout
plt.show()                            # Display plot

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species')  # Plot petal length vs width
plt.title("Petal Length vs Petal Width")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ---------------- Histograms ----------------

plt.figure(figsize=(8,6))
for col in df.columns[:-1]:           # Loop through all numeric columns (exclude 'Species')
    sns.histplot(df[col], kde=True, bins=20)  # Plot histogram with KDE curve

plt.legend(df.columns[:-1])           # Add legend for columns
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()

# ---------------- Boxplot (Outliers Detection) ----------------

plt.figure(figsize=(8,6))
sns.boxplot(data=df.select_dtypes(include='number'))  # Boxplot for numeric features only
plt.title("Boxplot for Outlier Detection")
plt.tight_layout()
plt.show()
```
